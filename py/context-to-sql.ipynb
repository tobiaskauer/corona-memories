```python
import pandas as pd
import sqlite3
import sqlalchemy
```

```python
context= pd.read_excel("./acaps_covid19_government_measures_dataset_0.xlsx", sheet_name=1, index_col="ID")

del context['ISO']
del context['REGION']
del context['ADMIN_LEVEL_NAME']
del context['PCODE']
del context['LOG_TYPE']
del context['TARGETED_POP_GROUP']
del context['NON_COMPLIANCE']
del context['SOURCE']
del context['SOURCE_TYPE']
del context['Alternative source']
del context['LINK']
del context['ENTRY_DATE']
```

```python
context['DATE'] = pd.to_datetime(context['DATE_IMPLEMENTED'], format='%Y/%m/%d')
context['DATE'] = context['DATE'].dt.date
del context['DATE_IMPLEMENTED']
context = context.rename(columns={"ID": "id", "COUNTRY": "country", "CATEGORY": "category", "MEASURE": "measure", "COMMENTS": "comment", "DATE": "date"})
context["weight"] = 1
context
```

```python
#SQLTE
sqlite = sqlite3.connect("../server/coronaMemories.sqlite")

context.to_sql(con=sqlite, name='contexts', if_exists='replace',dtype={"date": str(sqlalchemy.types.DateTime())})
```
